# 새로운 컬럼 - 위험도 만들기

# 주의 - 200000 미만
# 위험 - 200000~ 250000
# 경고 - 25~ 30만
# 최고 30만 초과

```python
import numpy as np
```

```python
warrning = [
    (df['확진자수'] < 200000) ,
    (df['확진자수'] > 200000) & (df['확진자수'] < 250000),
    (df['확진자수'] > 250000) & (df['확진자수'] < 300000),
    (df['확진자수'] > 300000)]
choicelist = ['주의', '위험', '경고', '최고']
df['위험도'] = np.select(warrning, choicelist, default='NON')
```

```python
def calc_danger(x):
    if x >300000:
        return '최고'
    elif x>250000:
        return '경고'
    elif x>200000:
        return '위험'
    else:
        return '주의'
```

```python
df['위험도2'] = df.apply(lambda x: calc_danger(x['확진자수']),axis=1)
```

```python
df
```

```python
df.groupby('위험도').mean().round(2) #mean()은 평균
```

```python
df.groupby('위험도')[['지역구','확진자수']].max()
```

```python
df.groupby('위험도')[['확진자수']].min()
```

```python
df.sort_values(by= ['확진자수'], ascending = False)
```
